## Step 3 — Create SparkSession

This is the boilerplate you will use in every PySpark job.
The S3 configuration is needed to read from your AWS bucket.

```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql.types import *

spark = (
    SparkSession
    .builder
    .appName("DEH-PySpark-Challenge")
    .config("spark.jars.packages",
            "org.apache.hadoop:hadoop-aws:3.3.4")
    .config("spark.hadoop.fs.s3a.impl",
            "org.apache.hadoop.fs.s3a.S3AFileSystem")
    .config("spark.hadoop.fs.s3a.aws.credentials.provider",
            "com.amazonaws.auth.EnvironmentVariableCredentialsProvider")
    .getOrCreate()
)

print(f"Spark version : {spark.version}")
print(f"App name      : {spark.sparkContext.appName}")
print(f"Master        : {spark.sparkContext.master}")
```

## Step 4 — Verify S3 Connection

Read the orders dataset from your S3 bucket to confirm everything is working.

```python
# Read orders from S3
orders = spark.read.csv(f"s3a://{BUCKET}/data/orders.csv", header=True)

print(f"Rows loaded: {orders.count()}")
orders.show(3)
```

## Step 5 — SparkSession Properties

These come up in interviews. Know what each one returns.

```python
# Access the SparkContext from the SparkSession
sc = spark.sparkContext

print(f"App name  : {sc.appName}")
print(f"Master    : {sc.master}")
print(f"Spark UI  : {sc.uiWebUrl}")

# Check a config value
print(f"App name from config: {spark.conf.get('spark.app.name')}")
```
